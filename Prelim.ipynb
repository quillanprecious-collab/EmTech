```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

circles=cv2.imread('/content/drive/MyDrive/EmTech/circles.png')
gray=cv2.cvtColor(circles, cv2.COLOR_BGR2GRAY)
blur=cv2.medianBlur(gray, 5)

hcircles=cv2.HoughCircles(
    blur,
    cv2.HOUGH_GRADIENT,
    dp=1,
    minDist=30,

    param1=10,
    param2=20,
    minRadius=20,
    maxRadius=80
)

cv2_imshow(hcircles)
```
